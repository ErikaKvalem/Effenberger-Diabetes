```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path_new = "level-4.csv"
df_new = pd.read_csv(file_path_new)

# Define the conditions to filter
conditions = ["Kontrolle", "pankreopriver Diabetes", "Diabetes mellitus Typ1"]

# Function to extract genus name (last part of taxonomy)
def extract_genus_name(taxonomy):
    """Extract the last part of a taxonomy string (genus name)."""
    return taxonomy.split(';')[-1].strip()

# Set up a figure with 3 subplots side by side
fig, axes = plt.subplots(1, 3, figsize=(24, 8))  # 1 row, 3 columns

# Loop through conditions and create a heatmap for each
for i, condition in enumerate(conditions):
    # Filter the dataset based on condition
    df_condition = df_new[df_new["condition"] == condition]

    # Select only numeric bacterial abundance data (excluding metadata)
    numeric_data = df_condition.select_dtypes(include=['number'])

    # Drop bacterial taxa (columns) that are completely missing or all zeros
    numeric_data = numeric_data.loc[:, (numeric_data != 0).sum() > 1]  # Keep only taxa that appear at least twice

    # Rename columns to retain only the genus name
    genus_names = {col: extract_genus_name(col) for col in numeric_data.columns}
    numeric_data.rename(columns=genus_names, inplace=True)

    # Remove columns where the genus name is empty ('')
    numeric_data = numeric_data.loc[:, numeric_data.columns != '']

    # Compute correlation matrix
    corr = numeric_data.corr()

    # Drop rows and columns that are entirely NaN (bacteria present in too few samples)
    corr.dropna(how="all", inplace=True)
    corr.dropna(axis=1, how="all", inplace=True)

    # Remove empty row and column labels from the correlation matrix
    corr = corr.loc[corr.index != '', corr.columns != '']

    # If there are no meaningful correlations left, skip this condition
    if corr.empty:
        print(f"Skipping {condition} due to insufficient data.")
        continue

    # Create a mask for the upper triangle
    mask = np.triu(np.ones_like(corr, dtype=bool))

    # Create the heatmap
    sns.heatmap(
        corr, mask=mask, cmap="coolwarm", annot=False, linewidths=0.5,
        vmin=-1, vmax=1, square=True, ax=axes[i]
    )

    # Set title for each heatmap
    axes[i].set_title(f"{condition}", fontsize=14)
    axes[i].tick_params(axis='x', rotation=90, labelsize=12)
    axes[i].tick_params(axis='y', rotation=0, labelsize=12)

# Adjust layout for better spacing
plt.tight_layout()

# Save the combined figure
#plt.savefig("./figures/correlogram_combined.svg", format="svg", dpi=300)
#plt.savefig("./figures/correlogram_combined.png", format="png", dpi=300)
#plt.savefig("./figures/correlogram_combined.pdf", format="pdf", dpi=300)

# Show the plot
plt.show()

print("Combined SVG, PNG, and PDF files have been saved successfully!")
```
